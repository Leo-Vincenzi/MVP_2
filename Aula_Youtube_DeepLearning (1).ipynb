```
# Configurando para não exibir os warnings no meio da execução do MVP
#
import warnings
warnings.filterwarnings("ignore")

# Fazendo os imports das bibliotecas
#
!pip install deepface # utiliza análise pelo método "convolutional neural network (CNN)"
import cv2 # ferramenta para trabalhar com imagens
from deepface import DeepFace # ferramenta para análise de atributos faciais (idade, gênero, emoção e raça)
import os # ferramenta para ler arquivos (e infos) do sistema operacional

# lendo todos os arquivos da pasta raiz
#
arquivos = os.listdir()

# para todos os arquivos "JPG"
#
for arquivo in arquivos:
    if "jpg" in arquivo:
	      # lendo a nossa imagem
        imagem = cv2.imread(arquivo)
	      # passando a imagem para a Deepface
        resultado = DeepFace.analyze(imagem, actions=("age", "gender", "emotion", "race"))
	      # mostrando os resultados da analise
        print(arquivo, resultado)
        print("oi")
```
