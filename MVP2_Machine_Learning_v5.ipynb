## 1 Definição do Problema
---
**Relevância**

A fraude em cartões de crédito é um problema global que causa enorme prejuízo financeiro. De acordo com o relatório *Card Fraud Losses Worldwide*, publicação comercial que cobre a indústria global de cartões de pagamento, perdas causadas por fraudes em cartões de crédito em todo o mundo chegarão a marca de **49** bilhões de dólares até **2030**. No Brasil, duas em cada dez pessoas já tiveram seus cartões de crédito fraudados.

Nesse viés, o advento do *Big Data* e a evolução de modelos mais robustos de *Machine Learning* permitiu o desenvolvimento de ferramentas que passaram a detectar e prevenir essas fraudes de maneira mais ágil e eficaz.

**Qual o objetivo desse MVP?**

O objetivo desse MVP é testar academicamente algoritmos e métodos de *Machine Learning* para classificar fraudes de cartão de crédito, tendo em vista uma determinada transação financeira, e ao mesmo tempo, minimizar as classificações incorretas de fraude.

Lembrando que o foco é mostrar, por meio desse MVP, o domínio e o conhecimento do assunto abordado.

**O que é fraude de cartão de crédito?**

Fraude de cartão de crédito ocorre quando alguém usa o cartão de crédito ou informações sigilosas da conta de outra pessoa, tanto em lojas físicas com on-line, para fazer compras não autorizadas, ou acessar fundos em dinheiro alheio.

**Como fazer descobrir a fraude?**

O problema da detecção de fraude de cartão de crédito inclui modelar transações anteriores de cartão de crédito ao achar padrões estatísticos, tendo como base o conhecimento daquelas que se revelaram uma fraude. Este modelo é então usado para identificar se uma nova transação é fraudulenta ou não.

**Sobre o *dataset* utilizado**

O *dataset* utilizado neste projeto é real, e contêm rgistro de transações de cartões de crédito efetuadas por clientes europeus, obtidas no mês setembro de 2013. Ele apresenta transações ocorridas durante dois dias (48 horas), por onde foram verificadas **492** fraudes em um total de em **284.807** transações financeiras (linhas).

A técnica PCA foi empregada porque reduz a dimensionalidade linear, por meio de uma transformação que cria um novo sistema de coordenadas, sendo comumente utilizada em análise exploratória de dados para o reconhecimento de padrões.

Além disso, não há metadados sobre os atributos originais, impossibilitando uma pré-análise ou o estudo qualitativo dos atributos.

Vamos agora verificar como é o desequilíbrio dos dados em relação ao atributo **"Class"**.


```python
# plotando um gráfico pizza mostrando a distribuição do atributo "Class"
counts = dataset['Class'].value_counts()
labels = ["Normal (0)", "Fraude (1)"] #labels = counts.index
explode = (0, 0)

# Definindo as linhas gerais do gráfico
plt.figure(figsize = (4, 5))
plt.pie(x=counts, labels=labels, explode=explode, autopct = '%1.2f%%', startangle=0, shadow=True, textprops={'fontsize': 14})
plt.title('Distribuição original do Atributo "Class"', weight = 'bold', color = '#2F0F5D', fontsize = 14)
plt.axis('equal')
plt.show()
```

Apenas **0,17%** (**492** ocorrências) das transações foram fraudulentas. Isso significa que temos dados altamente desequilibrados em relação ao atributo alvo **"Class"**. A justificativa é o baixo número de transações fraudulentas, quando comparadas às transações normais.

Qual o problema que ocorre quando há uma enorme diferença entre o número de classes positivas e classes negativas?

A resposta para essa pergunta é que, quando treinarmos o modelo para tal conjunto de dados, ele terá visto muito mais registros normais do que fraudulentos. Por isso será muito mais fácil para o modelo afirmar que não houve fraude, do que efetivamente descobrir uma faude (objetivo do projeto). E o resultado disso é que, ao tentar prever o alvo, o modelo sempre irá prever a classe em que aparece o maior número (99,83%), ou seja, resultado normal (Class=0).

Acurácia é uma das métricas mais utilizadas em problemas de classificação, mas em *datasets* com um atributo alvo muito desbalanceado como o nosso, a acurária é - em geral - muito alta e, por isso, enganosa.

Só que um alto valor de Acurácia não reflete a qualidade do modelo adotado, deixando de representar um indicativo válido, já que quando o modelo fizer a predição que a transação não é um fraude, a chance dele acertar é, no mínimo, 99,83%.

Portanto, para lidar com esse tipo de problema, podemos utilizar outras métricas como precisão, recall, *score* F1, matriz de confusão e curva ROC.

Se duas variáveis estiverem correlacionadas, podemos prever uma a partir da outra. Portanto, se duas características estão correlacionadas, o modelo necessita apenas de uma, pois a segunda não agrega informações adicionais.

Para esse caso, usaremos a correlação de **Pearson**, porque quando se trata de analisar dados e tomar decisões assertivas, a correlação de Pearson é considerada uma ferramenta muito útil.

O coeficiente de correlação de Pearson pode ter um intervalo de valores de +1 a -1. Um valor de 0 indica que não há associação entre as duas variáveis. Um valor maior que 0 indica uma associação positiva. Isto é, à medida que o valor de uma variável aumenta, o mesmo acontece com o valor da outra variável.

Precisamos definir um valor absoluto, digamos 0.5 (meio), como limite para seleção das variáveis. Se descobrirmos que as variáveis preditoras estão correlacionadas (cores mais escuras), podemos descartar a variável analisada com um valor de coeficiente de correlação inferior ao da variável alvo.

## 5 Preparação de Dados
---

A preparação de dados é uma tarefa que deve ser realizada com cautela, para evitar um conhecido problema chamado de *data leakage* (vazamento de dados). Isso ocorre durante a etapa de pré-processamento dos dados, quando há o “vazamento” dos dados de teste para os dados de treino, feita antes da divisão em treino e teste. Ou seja, é como se as informações de teste fossem reveladas durante o treinamento, causando uma estimativa final incorreta pelo modelo. Por isso o pré-processamento deve ser feito sempre após a divisão do *dataset* em treino e teste.

Há outras técnicas que podem ser utilizadas para minimizar a ocorrência de *data leakage*, como a utilização de *pipelines* e também a técnica de *cross-validation* (validação cruzada).

Iremos agora definir a lista dos atributos preditores (**X**) e a atributo alvo (**y**). Características categóricas, se houvessem, também poderiam ser definidas nesse ponto do projeto (no caso do nosso *dataset*, não há necessidade nesse momento).

```python
# separa atributos preditores do atributo alvo
X = dataset.drop('Class', axis=1)
y = dataset.Class
```

Nesse ponto, a partir do nosso *dataset* original, vamos definir e separar os subconjuntos de dados de treinamento, validação e teste, com o tamanho dos subconjuntos parametrizados com valore definidos no início do nosso MVP.

Dividiremos nossos dados em treinamento, validação e teste. Por causa do alto nível de desbalanceamento do *dataset*, utilizamos o algoritmo de divisão ***train_test_split*** no modo estratificado, que garante que cada divisão contenha aproximadamente o mesmo número ou distribuição de classes para cada divisão.

```python
# fazendo separação entre treino e teste

# holdout com estratificação
X_train, X_test, y_train, y_test = train_test_split(X, y,
      test_size=TEST_SIZE, random_state=RANDOM_STATE, shuffle=True, stratify=y)
```

## 6 Algoritmos Preditivos para Classificação

Um algoritmo preditivo em *Machine Learning* - de forma simplificada -  é uma função matemática que pode ser aplicada a um *Big Data*. O objetivo final desse tipo de algoritmo é evidenciar padrões capazes de apontar tendências e resultados.

### 6.1 Regressão Logística (LR)

A Regressão Logística (ou *Logistic Regression*) é um algoritmo de aprendizagem de máquina supervisionado que, apesar de ter a palavra "regressão" em seu nome, é utilizado em ML para classificação. A nomenclatura está relacionada ao fato de uma regressão logística ser construída a partir da aplicação de uma função (denominada "sigmoide") sobre a Regressão Linear.

A seguir rodaremos o algoritmo Regressão Logística usando ***LogisticRegression*** com nossa base de teste inicial (desbalanceada). Depois rodaremos nos dados de teste os algoritmos de *undesampling* e *oversampling*, avaliando-os ao final.

```python
# algoritmo LR - desbalanceado (leva 30 segundos)
clf_LR = LogisticRegression(random_state=RANDOM_STATE, max_iter=MAX_ROUNDS1) # definindo nosso classificador
clf_LR = clf_LR.fit(X_train, y_train) # ajustando o modelo
y_pred_LR = clf_LR.predict(X_test)    # fazendo a previsão com a nossa base de teste

# avaliando a métrica do LR
print(f"Acurácia : ({accuracy_score(y_test, y_pred_LR):.4f})")
print(f"Recall   : ({recall_score(y_test, y_pred_LR):.4f})")
print(f"Precisão : ({precision_score(y_test, y_pred_LR):.4f})")
print(confusion_matrix(y_test, y_pred_LR));
```

```python
# algoritmo LR - UnderSampler (US)
clf2_LR = LogisticRegression(random_state=RANDOM_STATE, max_iter=MAX_ROUNDS1) # definindo nosso clssificador
clf2_LR = clf2_LR.fit(X_res_US, y_res_US) # ajustando o modelo
y_pred2_LR = clf2_LR.predict(X_test);     # fazendo a previsão com a nossa base de teste

# avaliando a métrica do LR após rodar algoritmo UnderSampler (US)
print(f"Acurácia US: ({accuracy_score(y_test, y_pred2_LR):.4f})")
print(f"Recall   US: ({recall_score(y_test, y_pred2_LR):.4f})")
print(f"Precisão US: ({precision_score(y_test, y_pred2_LR):.4f})")
print(confusion_matrix(y_test, y_pred2_LR));
```

```python
# algoritmo LR - OverSampler (OS)
clf3_LR = LogisticRegression(random_state=RANDOM_STATE, max_iter=MAX_ROUNDS1) # definindo nosso clssificador
clf3_LR = clf3_LR.fit(X_res_OS, y_res_OS) # ajustando o modelo
y_pred3_LR = clf3_LR.predict(X_test)      # fazendo a previsão com a nossa base de teste

# avaliando a métrica do LR após rodar algoritmo OverSampler (OS)
print(f"Acurácia OS: ({accuracy_score(y_test, y_pred3_LR):.4f})")
print(f"Recall   OS: ({recall_score(y_test, y_pred3_LR):.4f})")
print(f"Precisão OS: ({precision_score(y_test, y_pred3_LR):.4f})")
print(confusion_matrix(y_test, y_pred3_LR));
```

Verificamos claramente, com o resultado acima, como a **Acurácia** é um indicador **ruim** para nossa análise, quando se trata de um *dataset* desbalanceado.

Como observamos no diagnóstico da matriz de confusão, em 56849 transações não foram achadas fraudes (VP-veradeiros positivos), enquanto foram achadas fraudes em 64 transações (VN-verdadeiros negativos). Então, se o modelo previsse todas as 100% das 56962 transações como normais, pelo menos em 56849 tentativas ele teria acertado!

E por mais que o modelo tivesse errado 64 vezes, isso acabaria sendo um numero muito pequeno quando se comparado com o total de acertos.

### 6.2 Árvore de Classificação (DT)

Árvore de Classificação (ou *Decision Tree*) é um algoritmo de aprendizado de máquina supervisionado com uma estrutura de simples compreensão, amplamente utilizado, que costuma apresentar bons resultados em suas previsões.

A Árvore de Classificação alia acurácia e interpretabilidade. Ela utiliza estruturas do tipo árvore compostas por nós, ramos e folhas. Os nós são as estruturas que armazenam as informações a respeito dos dados, onde são armazenados os testes e suas determinadas respostas em relação a um atributo.

O nó mais acima na árvore é denominado "nó raiz" (maior nível hierárquico), portanto pode-se visualizar a sua estrutura graficamente como uma árvore de cabeça para baixo, com a raiz no topo, ramos no meio e as folhas na base.

Os ramos representam os testes realizados nos nós, onde cada ramo retrata possíveis valores para a variável que está sendo analisada. E as folhas são o ponto final da árvore de decisão, representando os possíveis rótulos de classificação da variável alvo, e são definidas pelas regiões homogêneas.

A seguir rodaremos o algoritmo de Árvore de Classificação usando ***DecisionTreeClassifier*** com nossa base de teste inicial (desbalanceada). Depois rodaremos nos dados de teste os algoritmos de *undesampling* e *oversampling*, avaliando-os ao final.

### 6.3 KNN

O algoritmo *KNN* (*K-Nearest Neighbours*) é um algoritmo de classificação supervisionado que tem base na hipótese que instâncias similares são vizinhas uma das outras, de acordo com a métrica escolhida. Ele classifica novos pontos de dados tendo como base a “votação” da classe majoritária dos vizinhos mais próximos. ### 6.7 Erros Tipo I e Tipo II

Antes de terminarmos essa seção, sentimos a necessidade de melhor definir graficamente o que são Falsos Positivos (erros do *Tipo I*) e Falsos Negativos (do *Tipo II*).

Quando você rejeita algo que é verdadeiro, você está cometendo um erro do *Tipo I*. E quando não está rejeitando algo falso, você está cometendo um erro do *Tipo II*.

Na verdade, o que buscamos é de uma métrica que leve em conta, ao mesmo tempo, tanto a seletividade como a especificidade do método que estamos testando, de modo a minimizar, ao mesmo tempo, tanto os erros *Tipo I* como os *Tipo II*.

**Hipótese Nula** (**H0**) - A transação não é uma fraude.

**Hipótese Alternativa** (**H1**) - A transação é uma fraude.

**Erro tipo I** - Você rejeita a hipótese nula quando a hipótese nula é realmente verdadeira.

**Erro tipo II** - Você não consegue rejeitar a hipótese nula quando a hipótese alternativa é verdadeira.

**Custo do erro Tipo I** - Você presume erroneamente que a transação é uma fraude e uma transação verdadeira é rejeitada.

**Custo do erro Tipo II** - Você presume erroneamente que a transação não é uma fraude e uma transação fraudulenta é aceita.

## 7 Métodos Ensemble
---

A técnica de ML que combina múltiplos algoritmos - em busca de produzir um modelo preditivo mais robusto - é chamada de *ensemble* (ou comitês), e que possui métodos sequenciais (dependentes do modelo base) ou paralelos (independentes). Os métodos *ensemble* mais populares são os das classes *bagging, boosting e voting*.

Nesse momento precisamos definir alguns conceitos da análise estatística como sensibilidade e especificidade. Sensibilidade é a capacidade de o teste detectar corretamente resultados positivos, e por sua vez, especificidade é a capacidade de o teste detectar corretamente resultados negativos. Em análise preditiva, sensibilidade e especificidade são fundamentais para se aferir a acurácia dos modelos.

Um método bastante útil para visualizar de maneira conjunta a acurácia de determinado teste é a chamada Curva ROC (*Receiver Operating Characteristic*). A curva ROC permitem evidenciar, graficamente, os valores para os quais existe maior otimização da sensibilidade em função da especificidade, sendo comumente usada com classificadores binários.

A partir da curva ROC obtemos o valor AUC (*area under the ROC curve*), que é uma das métricas mais populares para a avaliação de um modelo de ML. AUC é considerada a melhor maneira de resumir o desempenho em um único número, já que ela mede a área sob a curva da Taxa de Verdadeiro Positivo (*Recall*) em função da Taxa de Falso Positivo.

Um modelo cujas previsões estão 100% erradas tem uma AUC de 0, enquanto um modelo cujas previsões são 100% corretas tem uma AUC de 1, e o limiar de predição na classificação é 0,5 (avaliador aleatório). Ou seja, acima desse limite, o algoritmo consegue ter efetividade para classificar uma classe.

Cada modelo ROC apresentará um determinado valor de AUC, o que irá auxiliar na escolha do melhor modelo, como apresentado no exemplo gráfico abaixo:

<img src="https://raw.githubusercontent.com/Leo-Vincenzi/PUC-MVP_2/main/MVP2_Curvas_ROC-AUC.jpg" width="600"/>

O ponto interessante do AUC é que a métrica não varia conforme a escala, já que não usa valores absolutos.

#### 7.5.2 Validação Cruzada

Vamos usar agora a **validação cruzada** para treinar e avaliar os modelos. Para tal usaremos validação cruzada **KFold**.

Os dados serão divididos em 5 partes (*folds*). Na primeira iteração a primeira parte será utilizada para teste e as partes restantes serão utilizadas para treinamento do modelo gerando uma métrica de avaliação. Na segunda iteração, a segunda parte será utilizada para teste enquanto as demais para treino.

Esse processo será repetido 5 vezes até que toda a base passe pelo processo de treino e teste, gerando uma métrica de avaliação média final para o modelo.

O valor de K (*NUMBER_KFOLDS*) é escolhido de forma que cada grupo de dados para treino e teste sejam grandes o suficiente para representarem estatisticamente o *dataset* original.

É possível encontrar na literatura a orientação para a utilização de K=5 ou K=10, valores encontrados através de experimentos, mas não é uma regra formal.

```python
# criando arrays e dataframes para armazenar os resultados (leva 5 minutos)
oof_preds = np.zeros(X_train.shape[0])
test_preds = np.zeros(X_test.shape[0])
feature_importance_df = pd.DataFrame()
n_fold = 0

for train_idx, valid_idx in kfold.split(X_train):
    train_x, train_y = X_train.iloc[train_idx],y_train.iloc[train_idx]
    valid_x, valid_y = X_train.iloc[valid_idx],y_train.iloc[valid_idx]

    evals_results = {}
    # definindo hiperparâmetros
    model =  LGBMClassifier(
                  nthread=-1,
                  n_estimators=2000,
                  learning_rate=0.01,
                  num_leaves=80,
                  colsample_bytree=0.98,
                  subsample=0.78,
                  reg_alpha=0.04,
                  reg_lambda=0.073,
                  subsample_for_bin=50,
                  boosting_type='gbdt',
                  is_unbalance=False,
                  min_split_gain=0.025,
                  min_child_weight=40,
                  min_child_samples=510,
                  objective='binary',
                  metric='rec',       # métrica escolhida
                  silent=-1,
                  verbose=-1,
                  feval=None)

    # inicializando e ajustando o modelo
    model.fit(train_x, train_y, eval_set=[(train_x, train_y), (valid_x, valid_y)], eval_metric= 'rec')

    oof_preds[valid_idx] = model.predict_proba(valid_x, num_iteration=model.best_iteration_)[:, 1]
    test_preds += model.predict_proba(X_test, num_iteration=model.best_iteration_)[:, 1] / kfold.n_splits

    fold_importance_df = pd.DataFrame()
    fold_importance_df["feature"] = predictors
    fold_importance_df["importance"] = model.feature_importances_
    fold_importance_df["fold"] = n_fold + 1

    feature_importance_df = pd.concat([feature_importance_df, fold_importance_df], axis=0)
    print('Parte%2d - AUC: %.4f' % (n_fold + 1, roc_auc_score(valid_y, oof_preds[valid_idx])))
    del model, train_x, train_y, valid_x, valid_y
    gc.collect()
    n_fold = n_fold + 1
train_auc_score = roc_auc_score(y_train, oof_preds)

# print('A Pontuação AUC final é %.4f' % train_auc_score)
```

Vamos calcular a área sobre a curva **ROC** com LightGBM calculando **AUC**, usando a validação cruzada.

```python
print(f"A pontuação AUC final do método LightGBM com validação cruzada (K={NUMBER_KFOLDS}) para a previsão dos dados do teste foi de {train_auc_score:.4f}")
```

### 7.6 Voting Classifier

*Ensemble* do tipo *voting* é um dos métodos mais simples e *VotingClassifier* (Classificação por Votação) é o modelo que avaliaremos desse tipo. A abordagem *voting* consiste em combinar as previsões de dois ou mais algoritmos independentes, chamados de classificadores base, para escolher uma predição final, feita por votação majoritária.

O método *voting* geralmente tem melhor resultado do que o melhor preditor individual da lista de elementos usada no *ensemble*. E mesmo que todos os preditores escolhidos para a votação tenham perfomances iniciais relativamente baixas, o *ensemble* do tipo *voting* ainda pode conseguir um bom resultado final, desde que os algoritmos escolhidos para participar da votação sejam suficientemente numerosos e diversos.

Nesse *ensemble* abaixo faremos uma votação entre 3 modelos de algoritmo: LR, DT e NB, baseado no resultado da comparação de algoritmos feita na Seção 6.

```python
# separando em bases de treino e teste (holdout) (leva 3 minutos)
array = dataset.values

# separando os dados em folds
kfold = KFold(NUMBER_KFOLDS, shuffle=True, random_state = RANDOM_STATE)

# criando os modelos para uso do VotingClassifier
estimators = []
model_1 = LogisticRegression(max_iter = MAX_ROUNDS1)
estimators.append(('logistic', model_1))
model_2 = DecisionTreeClassifier()
estimators.append(('cart', model_2))
model_3 = GaussianNB()
estimators.append(('NB', model_3))

# criando o modelo ensemble
model = VotingClassifier(estimators)

# fazendo o cross validation
cv_results = cross_val_score(model, X, y, cv = kfold, scoring=SCORING)

# mostrando o resultado
print(f"A pontuação AUC obtida com o método VotingClassifier foi de {cv_results.mean():.4f}")
```

```python
# Resultados Voting - será apresentado na seção 7.7 (leva minutos)
name = 'Voting'
results.append(cv_results) # inclui resultado no gráfico
names.append(name) # inclui nome do método no gráfico
msg = "%s: %.4f (%.4f)" % (name, cv_results.mean(), cv_results.std())
print(msg)
```

Estudamos nesse MVP o funcionamento de alguns algoritmos de *Machine Learning* usados para classificação: Regressão Logística, Árvore de Decisão, KNN, *Naive Bayes* e SVM. E verificamos como, em alguns casos, o desequilíbrio de classes afetou o desempenho dos algoritmos.

As métricas de Precisão e *Recall* forneceram uma compreensão mais profunda ao nosso problema, quando comparado à Acurácia. A Precisão se concentra em minimizar os Falsos Positivos (FP), enquanto o *Recall* enfatiza na maximização dos Verdadeiros Positivos (VP). E ao examinarmos ambas as métricas, podemos encontrar um equilíbrio que se alinhe com nossas prioridades.

Dependendo do contexto, podemos priorizar a Precisão em detrimento do *Recall*, ou vice-versa. A dificuldade está em achar um modelo que tenha uma alta predição de fraudes, mantendo baixos os Falsos Positivos (FP) e Falsos Negativos (FN).

Em seguida, observamos alguns métodos *ensemble*. Verificamos na prática que alguns são mais lentos, e com outros obtivemos resultados mais ágeis.

*RandomForrest, AdaBoost, CatBoost, XGBoost, LightGBM e Voting* oferecem vários hiperparâmetros, que possibilitam inúmeros ajustes para otimizar o desempenho do modelo, e que costumam incluir, por exemplo, taxa de aprendizagem, profundidade da árvore, número de árvores, termos de regularização, etc.

Ainda pensando na finalidade desse estudo, para os métodos *ensemble* usamos, como métrica de avaliação a Matriz de Confusão, curva ROC e indicador AOC. A pontuação AOC representa a área debaxio da curva ROC, então, na prática, o que estamos fazendo é - matematicamente - o **cálculo da integral** de uma função no intervalo desejado.

Depois de testar todos esses modelos e algumas das suas opções, a pergunta que fica é: ***qual o melhor método*** ?

Existem os mais populares, mas não temos como afirmar positivamente que um modelo específico será sempre a melhor escolha para todos os casos.

Trabalhando com *Machine Learning* e *Deep Learning* certamente iremos lidar com situações diferentes em ambientes diferentes. E para cada situação específica, um modelo de algoritmo pode apresentar melhores resultados do que o outro, ao lidar melhor com aprendizado supervisionado ou não, atributos com características categóricas ou de cardinalidade, disponibilidade maior ou menor de tempo e recursos computacionais, acurácia, *recall*, precisão, escalabilidade, etc.

A **Acurácia** normalmente é uma boa indicação de perfomance de um modelo. Porém, em algumas situações, ela pode ser enganosa.

Isso ficou claro no exemplo do nosso modelo de identificação de fraudes em cartões de crédito. Verificamos por exemplo, que na validação de um modelo de regressão logística, obtivemos acurácia de 0.9991, e que mesmo assim, ele deixa passar 34 de um total de 98 fraudes!

Ao acrescentarmos o uso de algoritmos de balanceamento em nosso modelo de regrassão logística, podemos melhorar o número de predição de fraudadores para 88 ou 89, mas ainda estaremos deixando de prever mais de 10% de clientes desonestos, e ainda aumentando absurdamente o número de Falsos Positivos: de 15 clientes iniciais para cerca de 2000 clientes.

Se voltarmos a atenção para a métrica **Precisão**, verificamos que ela pode ser usada para melhor indicar a perfomance de um modelo em situações que os Falsos Positivos são considerados mais prejudiciais que os Falsos Negativos. Porém, esse não é bem o nosso caso.

Finalmente, o ***Recall*** é a métrica de avaliação de performance que se adequa melhor às situações em que os Falsos Negativos são considerados mais prejudiciais do que os Falsos Positivos. Por isso nos fornece a melhor medida para o problema específico em estudo.

Nosso projeto é um bom exemplo de busca de um modelo que prioriza encontrar os fraudadores, mesmo que ele eventualmente classifique alguns clientes honestos como não honestos. Por esse motivo o *Recall* foi uma das nossas métricas escolhidas para esse MVP (parâmetro SCORING).

O *final tunning* do nosso modelo irá depender do que foi decidido na reunião inicial que tivemos com o cliente que gerencia a bandeira de cartão de crédito, ou seja, a empresa responsável que está por trás de processar as transações financeiras.

Ela é quem vai avaliar até que ponto se deseja minimizar as fraudes, no caso de uma bandeira voltada para o público de maior renda, em que os valores das transações costumam ser mais elevados, ou até que ponto a empresa gostaria de evitar a importunação causada aos clientes por um Falso Positivo durante uma transação.

Por outro lado, podemos ter o caso de uma bandeira mais focada para um público popular, que lida em média com compras de valores mais baixos, e que os prejuízos causados por Falsos Negativos não são tão elevados.

Um outro aspecto que se deve levar em conta é se a bandeira possui uma rede de canais de atendimento ao cliente robusta, capaz de atender a um grande número de clientes honestos e que tiveram suas transações negadas. Também devemos consider a abrangência da rede credenciada, já que num país com dimensões como a do Brasil, por exemplo, isso poderá influenciar no tempo de resposta em que o cliente levará para conseguir em contato com a bandeira e confirmar seus dados, numa eventual transação negada.

O resultado a que chegamos é que cada métrica tem sua peculiaridade, e que a escolha deve - principalmente - levar em consideração o problema abordado. Também não existe modelo com solução 100%. O que podemos fazer, em última análise, é encontrar um equilíbrio que permita escolher uma métrica e um modelo que consiga se adequar melhor às verdadeiras necessidades do cliente no nosso projeto de negócios.

Sempre há espaço para melhorias, como por exemplo experimentar diferentes hiperparâmetros nos algoritmos, escolhendo o tipo de balanceamento dos dados ou na escolha seletiva dos atributos preditores, mas a conclusão que tiramos é que modelos de aprendizado de máquina já desempenham - e cada vez mais desempenharão - um papel crucial na abordagem do problema abordado por esse MVP.
